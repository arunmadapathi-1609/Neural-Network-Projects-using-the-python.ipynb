```python
import numpy as np
import pandas as pd
import sklearn.datasets
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_dataset=sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
data_frame=pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
```

```python
data_frame['label']=breast_cancer_dataset.target
```

```python
x=data_frame.drop(columns='label', axis=1)
y=data_frame['label']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
#standardizing the data

from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_train_std=scaler.fit_transform(x_train) #the fit function is only used for the training data
x_test_std=scaler.transform(x_test)
```

Building The Neural Network

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers in the Neural Network

model=keras.Sequential([
                        keras.layers.Flatten(input_shape=(30,)),
                        keras.layers.Dense(20, activation='relu'),
                        keras.layers.Dense(2, activation='sigmoid')

])
```

```python
# sompiling the Neural Network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
#training the model
history=model.fit(x_train_std,y_train, validation_split=0.1,epochs=10) #epochs is the no of times the model goes through the data
```

```python
# visulazing accurcay and loss

plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model loss')

plt.xlabel('epoch')
plt.ylabel('loss')


plt.legend(['training data', 'validation data'], loc='lower right')
```

```python
# accuracy of the model on test data

loss,accuracy =model.evaluate(x_test_std,y_test)
print(accuracy)
```

```python
print(x_test_std.shape)
print(x_test_std[0]) # the first data in the test data
```

```python
y_pred=model.predict(x_test_std) # model predict gives the probability of each class for the data point
```
